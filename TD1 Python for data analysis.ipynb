# Passons a un dataset pris sur internet

```python
# Trouver un dataset
#analyser 
#3 questions et 3 graphes
corona = pd.read_csv("coronavirus_evolution_par_jour_et_par_pays.csv",sep=";")
#liens : https://www.data.gouv.fr/fr/datasets/coronavirus-covid19-evolution-par-pays-et-dans-le-monde-maj-quotidienne/

#Information dans les fichiers (csv et json)
#Pays
#Nombre de cas total
#Nombre de décès total
#Nombre de guérison total
#Taux de décès (en pourcentage) => (nbDeces / nbCas) * 100
#Taux de guérison (en pourcentage) => (nbGuerison / nbCas) * 100
#Taux d'infection (personnes encore infectées, non décédés ni guéris) (en pourcentage) => 100 - Taux de deces - Taux de guerison
```

```python
corona.head()
```

```python
#Nombre de cas de le monde en fonction des jours
corona[:].groupby("Date").Infections.sum()[:]
```

```python
#Nombre de cas de le monde par jour
corona[:].groupby("Date").Infections.sum()[:].diff(1)
#300 000 nouveaux cas dans le monde par jour => le dernier jour etait probablement incomplet car pris avant l'enregistrement des données
```

```python
#Pays les plus touchés
corona[corona['Date'] == "2020-09-10"].sort_values(by="Infections")
```

```python
#evolution du nombre total d'infections du corona en France sur les 10 derniers jours
corona[(corona.Pays=="France")].groupby("Date").Infections.sum()[-10:]
```

```python
#evolution du nombre total de cas de coronavirus en france sur la période 2020-01-24 au 2020-09-17
corona[(corona.Pays=="France")].groupby("Date").Infections.sum()[:].plot()
```

```python
#mon dataset ne possede pas le nombre de cas par jour alors je vais le calculer pour la France
nb_cas_total_jour = corona[(corona.Pays=="France")].groupby("Date").Infections.sum()[:]
nb_cas_jour = nb_cas_total_jour.diff(1);
print(nb_cas_jour)
```

```python
#Evolution du nombre de cas par jour en France
nb_cas_jour.plot()
#on peut voir qu'il y a un jour avec une valeure négative, probablement une valeur à retirer
```

```python
#Essayons maintenant de comparer par rapport à d'autres pays
#Les pays ayant plus de 100 000 cas
pays_plus_100k = corona[corona['Infections'] > 100000]
pays_plus_100k['Pays'].drop_duplicates().count() #equivalenet de Unique value en SQL
```

```python
#Taux d'infection comparaison entre la France et l'Allemagne
france = corona[(corona.Pays=="France")].groupby("Date").TauxInfection.sum()[:]
allemagne = corona[(corona.Pays=="Allemagne")].groupby("Date").TauxInfection.sum()[:]
france_allemagne = pd.merge(france, allemagne, left_on = ['Date'], right_on = ['Date'])  #Equivalent de join en SQL
```

```python
france_allemagne.plot()
```

```python
#Dans le monde qu'elles ont été le nombre maximu de cas par pays et depuis combien de jours ils ont le corona
#values = valeur auquel on veut appliquer nos aggfunc
#index = axe des x
#collums = axe des y
#aggfunc = function de calcul sur nos values => max, min, np.mean, np.std, sum , len
table = pd.pivot_table(corona, values='Infections', index=['Date'],columns='Pays',aggfunc=max) 
```

```python
table.plot(title="Evolution du corona virus dans le monde")
```

```python
#ameliorons exemple precedent en prenons que les 3 pays les plus touchés

corona_3_pays = corona[(corona.Pays=="Brésil") | (corona.Pays=="Inde") | (corona.Pays=="États-Unis")]
table_3_pays = pd.pivot_table(corona_3_pays, values='Infections', index=['Date'],columns='Pays',aggfunc=max) 
table_3_pays.plot(title="Evolution du corona virus dans le monde")
```
